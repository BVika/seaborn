```python
#импортируем библиотеки seaborn и matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#графики будут строиться с использованием настроек из seaborn
sns.set()
```

```python
#Загрузим данные из файла curs (1).csv, сохраним их в датафрейм 
import pandas as pd
ch = pd.read_csv("pingvin.csv")
ch.dropna(inplace = True)
ch.head()
```

**Показатели:**

* `species	`: Вид;
* `island	`: Место обитания;
* `bill_length_mm`: Длина клюва;
* `bill_depth_mm`: Высота клюва;
* `flipper_length_mm`: Длина лап;
* `body_mass_g`:Масса тела;
* `sex`: пол;


### Гистограмма и график плотности распределения

```python
#Построим  гистограмму для body_mass_g
sns.distplot(ch["body_mass_g"]) 
```

```python
#уберём сглаженный график плотности и изменим цвет гистограммы.
sns.distplot(ch["body_mass_g"], kde = False, 
             color = "r")
```

```python
#kde отвечает за наличие или отсутствие сглаженного графика плотности
#bins числа столбцов
#hist_kws перечень аргументов для построения гистограммы
# alpha – прозрачность (1 - нет прозрачности)
# edgecolor – цвет границ столбцов

sns.distplot(ch["body_mass_g"], 
             kde = False, 
             color = "r", 
             bins = 15, 
             hist_kws = {'alpha' : 0.6, 
                        'edgecolor' : 'black'})
```

```python
sns.distplot(ch["body_mass_g"], 
             color = "r", 
             bins = 10, 
             hist_kws = {'alpha' : 0.6, 
                        'edgecolor' : 'black'},
             kde_kws = {'color' : 'b'})
```

```python
#добавим заголовок графика и подписи к осям
sns.distplot(ch["body_mass_g"], 
             color = "r", 
             bins = 10, 
             hist_kws = {'alpha' : 0.75, 
                        'edgecolor' : 'black'},
             kde_kws = {'color' : 'b'})
plt.xlabel("body mass, g")
plt.ylabel("Counts")
plt.title("Вес особи")
```

```python
#Добавим горизонтальную линию
sns.distplot(ch["body_mass_g"], 
             color = "r", 
             bins = 10, 
             hist_kws = {'alpha' : 0.75, 
                        'edgecolor' : 'black'},
             kde_kws = {'color' : 'b'})
plt.xlabel("body mass, g")
plt.ylabel("Counts")
plt.title("Вес особи")

# метод hlines
plt.hlines(y = 0.0006, xmin = 10, xmax = 7000, 
          color = "red", linestyles = "dashed")
```

```python
sns.distplot(ch["body_mass_g"], 
             color = "r", 
             bins = 10, 
             hist_kws = {'alpha' : 0.75, 
                        'edgecolor' : 'black'},
             kde_kws = {'color' : 'b'})
plt.xlabel("body mass, g")
plt.ylabel("Counts")
plt.title("Вес особи")

# метод vlines
# вертикальная линия отмечает среднее значение по столбцу
plt.vlines(x = ch["body_mass_g"].mean(), ymin = 0, ymax = 0.0009, 
          color = "black", linestyles = "dashed")
```

```python
#сглаженный график плотности
sns.kdeplot(ch["body_mass_g"], shade = True)
plt.xlabel("body mass, g")
plt.ylabel("Counts")
plt.title("Вес особи")
```

```python
# «выключаем» гистограмму
# «включаем» rugs

sns.distplot(ch["body_mass_g"], 
             hist = False, 
             rug = True)
plt.xlabel("body mass, g")
plt.ylabel("Counts")
plt.title("Вес особи")
```

```python
# функция FacetGrid создает сетку
# margin_titles – наличие подписей по строкам и столбцам

g = sns.FacetGrid(ch, row="sex", 
                  col="species", 
                  margin_titles=True)
g
```

```python
g = sns.FacetGrid(ch, row="sex", 
                  col="species", 
                  margin_titles=True)
g.map(plt.hist, "island", color="steelblue")
```

### Столбиковая диаграмма
Построим столбиковую диаграмму (bar plot или bar chart)

```python
#данные не подходят для построения столбиковой диаграммы
tab = ch["species"].value_counts()
tab
```

```python
x = tab.index
y = tab.values
x, y
```

```python
sns.set_style("whitegrid")
sns.barplot(x, y)
```

```python
res = ch.groupby(['species', 'sex'], as_index=False).count()
res
```

```python
ch.groupby(['species', 'sex']).count() 
```

```python
freq = res[['species', 'sex', 'body_mass_g']]
freq
```

```python
sns.catplot(x = "species", 
            y = "body_mass_g", 
            hue = "sex", 
            data=freq,
            kind="bar",
            palette = "Pastel1")
plt.ylabel("Counts")
plt.xlabel("species")
```

```python
g = sns.FacetGrid(data=ch,
                  col='island',    # одна колонка - один остров
                  row='sex',       # пол особи в отдельной строке
                  hue='species')   # цветом выделяем вид пингвинов

g.map(plt.scatter,             # указываем нужную функцию matplotlib
      'body_mass_g',           # значения по оси X
      'flipper_length_mm',     # значения по оси Y
      s=60, alpha=0.3);        # ключевые аргументы функции 

# Устанавливаем подписи осей:
g.set_axis_labels('Масса тела (г)',
                  'Длина крыла (мм)');

# Устанавливаем общий заголовок
g.fig.suptitle('Зависимость длины крыльев от массы тела',
               y = 1.03,
               fontsize = 20);
```
